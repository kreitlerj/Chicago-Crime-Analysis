```python
data_with_month.columns = ['Date_to_drop',
                           'Date',
                           'Year',
                           'Month',
                           'Time',
                           'Primary Type',
                           'Description',
                           'Ward',
                           'Latitude',
                           'Longitude'] 
```

```python
#data_with_month
```

```python
cleaned_data = data_with_month.drop(['Date_to_drop', 'Description'], axis=1) 



```

```python
#cleaned_data
```

```python
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'OTHER OFFENSE']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'PUBLIC PEACE VIOLATION']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'INTERFERENCE WITH PUBLIC OFFICER']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'GAMBLING']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'PUBLIC INDECENCY']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'OBSCENITY'] 
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'INTIMIDATION'] 
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'DECEPTIVE PRACTICE'] 
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'STALKING']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'LIQUOR LAW VIOLATION']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'OFFENSE INVOLVING CHILDREN']
```

```python
#cleaned_data
```

```python
#combining primary types
cleaned_data.loc[cleaned_data['Primary Type'] == 'OTHER NARCOTIC VIOLATION', 'Primary Type'] = 'NARCOTICS'
cleaned_data.loc[cleaned_data['Primary Type'] == 'ASSAULT', 'Primary Type'] = 'ASSAULT/BATTERY'
cleaned_data.loc[cleaned_data['Primary Type'] == 'BATTERY', 'Primary Type'] = 'ASSAULT/BATTERY'
cleaned_data.loc[cleaned_data['Primary Type'] == 'CRIM SEXUAL ASSAULT', 'Primary Type'] = 'SEX OFFENSE'

cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'NARCOTICS']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'CRIMINAL DAMAGE']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'BURGLARY']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'THEFT']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'PROSTITUTION']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'CRIMINAL TRESPASS']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'MOTOR VEHICLE THEFT']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'WEAPONS VIOLATION']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'CONCEALED CARRY LICENSE VIOLATION']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'NON-CRIMINAL']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'HUMAN TRAFFICKING']
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'NON-CRIMINAL (SUBJECT SPECIFIED)'] 
cleaned_data = cleaned_data[cleaned_data['Primary Type'] != 'ARSON']  

```

```python
cleaned_data['Primary Type'].value_counts()
```

```python
cleaned_data = cleaned_data.drop(['Latitude','Longitude'], axis=1)
```

```python
cleaned_data
```

```python
Chicago_Crime_2017 = cleaned_data[cleaned_data['Year'] == 2017]
Chicago_Crime_2018 = cleaned_data[cleaned_data['Year'] == 2018] 
```

```python
#push to sqlite database 
Chicago_Crime_2017.to_sql('Crime_2017', con=engine,
                   if_exists='replace')
Chicago_Crime_2018.to_sql("Crime_2018", con=engine,
                         if_exists='replace')
    
    
    
```
